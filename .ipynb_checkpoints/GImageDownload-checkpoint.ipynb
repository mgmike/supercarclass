# Download images

```python
#Make sure to use following gid instead of main one:
#pip3 install git+https://github.com/Joeclinton1/google-images-download.git
from google_images_download import google_images_download as gid
import os
import shutil
```

```python
def download_images(directory, model, limit):
    args = {'keywords':model, 'limit':limit, 'format':'jpg', 'output_directory':directory, 'image_directory':model,'size':'medium','suffix_keywords':'exterior'}
    response = gid.googleimagesdownload()
    paths = response.download(args)
    #print(paths)
#     for path in paths[0][model + ' exterior']:
#         split = path.split('.')
#         if split[len(split) - 1] != 'jpg':
#             os.remove(path)
    
    
    
def move_images(directory, model):
    dir_list = os.listdir(directory + '/train_test/' + model)
    print(dir_list)
    if not os.path.exists(directory + '/train/' + model):
        os.mkdir(directory + '/train/' + model)
    if not os.path.exists(directory + '/test/' + model):
        os.mkdir(directory + '/test/' + model)
    
    for i, filename in enumerate(dir_list):
        if i < int(len(dir_list) * train_ratio):
            shutil.copyfile(directory + '/train_test/' + model + '/' + filename, directory + '/train/' + model + '/' + filename)
        else:
            shutil.copyfile(directory + '/train_test/' + model + '/' + filename, directory + '/test/' + model + '/' + filename)
    
def rename_images(directory, model, data_list):
    directory_model = directory + '/' + model
    for filename in os.listdir(directory_model):
        last_file_number = int(data_list[len(data_list) - 1].split(',')[0].split('.')[0])
        new_name = '%0*d' %(5, last_file_number + 1) + '.jpg'
        os.rename(directory_model + '/' + filename, directory_model + '/' + new_name)
        data_list.append(new_name + ',0')
        
def redistribute_images(directory, model, train_test_ratio=0.8):
    train_list = os.listdir(directory + '/train/' + model)
    test_list = os.listdir(directory + '/test/' + model)
    total = len(train_list) + len(test_list)
    while len(train_list) < (total * train_test_ratio):
        train_list.append(test_list.pop(0))
    for filename in train_list:
        if os.path.exists(directory + '/test/' + model + '/' + filename):
            shutil.move(directory + '/test/' + model + '/'  + filename, directory + '/train/' + model + '/'  + filename)
            
def remove_nonjpeg(directory, model):
    train_list = os.listdir(directory + '/train/' + model)
    test_list = os.listdir(directory + '/test/' + model)
    for filename in train_list:
        if '.jpg' not in filename:
            os.remove(directory + '/train/' + model + '/' + filename)
            removed += 1
    for filename in test_list:
        if '.jpg' not in filename:
            os.remove(directory + '/test/' + model + '/' + filename)
            removed += 1
```

```python
cwd = os.getcwd()
directory = cwd + '/new_cars'
limit = 100
train_ratio = 0.8
model = 'Mclaren 650S'

anno_train = []
anno_test = []
names = []
cars_tba = []

for line in open(cwd + '/anno_train.csv'):
    anno_train.append(line)
for line in open(cwd + '/anno_test.csv'):
    anno_test.append(line)
for line in open(cwd + '/names.csv'):
    names.append(line)
for line in open(cwd + '/new_names.csv'):
    cars_tba.append(line.strip('\n'))
```

```python
for model in cars_tba:
    if not os.path.exists(directory + '/train_test/' + model):
        download_images(directory + '/train_test', model, limit)
```

```python
for model in cars_tba:
    move_images(directory, model)
    rename_images(directory + '/train', model, anno_train)
    rename_images(directory + '/test' , model, anno_test)

    names.append(model)
```

```python
for model in cars_tba:
    redistribute_images(directory, model)
```

```python
removed = 0
for model in cars_tba:
    remove_nonjpeg(directory, model)
```
